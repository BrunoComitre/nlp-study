# SentiLex-PT01

SentiLex-PT01 is a lexicon for the Portuguese language that is available for research and development.

SentiLex-PT01 is especially useful for opinion-mining applications involving Portuguese, in particular to detect and classify feelings and opinions about human entities.

It is available in two separate .txt files:

1. SentiLex-lem-PT01.txt:

The entries are represented by a lemma (conventionally the singular masculine form) and its speech part (Adj), followed by the sentiment attributes described below:

- polarity (POL), which may be positive (1), negative (-1) or neutral (0);
- polarity target (TG), which corresponds to a human subject (HUM);
- ANOT, which was executed manually (MAN) or automatically, by the Judgment Analysis Lexical Classifier (JALC), developed by the project team.

Example:
- bonito.POS = Adj; POL = 1; TG = HUM; ANOT = MAN
- off. POS = Adj; POL = -1; TG = HUM; ANOT = JALC

2. SentiLex-flex-PT01.txt:

Adjectives are inflected in gender (G) and number (N), which are associated with their corresponding lemma. In addition to the linguistic information described in the lemma dictionary, in SentiLex-flex-PT01.txt, each adjective is classified as masculine (m) or feminine (f) and singular (s) or plural (p).

Example:
- pretty pretty. POS = Adj; GN = fs; POL = 1; TG = HUM; ANOT = MAN
- Beautiful, beautiful. POS = Adj; GN = fp; POL = 1; TG = HUM; ANOT = MAN

<br/>

**Polarity distribution:**

- 3,494 adjectives with NEGATIVE polarity
- 1,243 have POSITIVE polarity
- 1,584 have NEUTRAL polarity

<br/>

**Polarity assignment:**

- 3,585 adjectives were manually labeled by a linguist.
- 2,736 adjectives were automatically labeled.

The JALC algorithm used has an overall precision of 67%.

It is more accurate in the classification of negative adjectives (82% accuracy) than positive adjectives (67%).

The most problematic cases involve neutral polarity, which is, on average, correctly attributed only in 45% of cases.

# Applying SentilexPt

**Open the file and read the information:**

```python
sentilexpt = open('SentiLex-flex-PT01.txt','r', encoding='utf-8')
```

**Creating dictionary to receive information:**

```python
dic_palavra_polaridade = {}
```

**Attributing each word to a Polarity:**

```python
for i in sentilexpt.readlines():
    pos_ponto = i.find('.')
    palavra = (i[:pos_ponto])
    palavra = palavra.split(',')[0]
    #print(palavra)
    pol_pos = i.find('POL')
    polaridade = (i[pol_pos+4:pol_pos+6]).replace(';', '')
    dic_palavra_polaridade[palavra] = polaridade
```

In this code the words and polarity were separated by sweeping the strings. After that, they were stored in a dictionary, which contains the word and the feeling attributed to the word.

**Visualize:**

```python
print(dic_palavra_polaridade)
```

**Example:**

```python
print (dic_palavra_polaridade.get('abusivo'))
```

To know the feeling of a word, just consult the dictionary, as in the example above.

**Assigning Sentiment Score:**

```python
def Score_sentimento(frase):
    frase = frase.lower()
    l_sentimento = []
    for p in frase.split():
        l_sentimento.append(int(dic_palavra_polaridade.get(p, 0)))
        #print("palavra=%s polaridade=%s" % (p,int(dic_palavra_polaridade.get(p, 0))))
    
    score = sum(l_sentimento)
    if score > 0:
        return 'Green - Positivo, Score:{}'.format(score)
    elif score == 0:
        return 'Yellow - Neutro, Score:{}'.format(score)
    else:
        return 'Red - Negativo, Score:{}'.format(score)
```

For the Analysis of Feelings in sentences, the process has to be done word-for-word processing.

This function picks up the dictionary made in the previous function and it will return me to query the feeling and each word of a sentence and add them all together.

If the result is:
- 1 or greater than 1 = POSITIVE
- 0 = NEUTRAL
- less than 0 = NEGATIVE

**Testing:**

```python
Score_sentimento('Eu estou muito feliz hoje, porém, triste com a politica')
```

As this phrase has positive and negative words, ie words with opposite feelings, the sum was 0 and the function returned 0.

```python
dic_palavra_polaridade.get('triste')
```

# Polaridade do Sentimento

Example of Lexical Analysis in a .txt file, containing sentences in Portuguese.

**Example:**

```python
count = 1
sentiment = {'documents':[]}
```

```python
f = open("sentence.txt", "r", encoding='utf-8')
```

```python
for frase in f:
    frase = frase.replace('\r','').replace('\n','')
    print('id:%d\t %s' % (count,frase))
    score = Score_sentimento(frase)
    sentiment['documents'].append({'id':str(count),'language': 'pt', 'text': frase, 'score': score })
    count += 1
```

```python
for n in sentiment['documents']:
    print("id=>%s \tsentiment=%s" % (n['id'], n['score']))
```

This method is the fastest, but what more must be careful, because with the low amount of people contributing like Lexico, we can have a higher error rate than the others.

# Using Naive Bayes Analyzer

**Imports and Parameters:**

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
```

**Open the file and read the information:**

```python
f = open("sentence.txt", "r", encoding='utf-8')
```

**Example:**

```python
count = 1
sentiment = {'documents':[]}
```

**Using TextBlob to add translation:**

```python
for frase in f:
    frase = frase.replace("\r","").replace("\n","")
    print("id:%d\t %s" % (count,frase))
    frase = TextBlob(frase)
    traducao = str(TextBlob(str(frase.translate(to='en'))))
    blob = TextBlob(traducao)
    sentiment['documents'].append({'id':str(count),'language': 'pt', 'text': frase, 'score': blob.sentiment.polarity })
    count += 1
```

**Result:**

```python
for n in sentiment['documents']:
        if n['score'] <= -0.1 :
                print("id=>%s \tsentiment=Red \tscore=>%s" % (n['id'], n['score']))

        if n['score'] >= 0.0 and n['score'] <= 0.65:
                print("id=>%s \tsentiment=Yellow \tscore=>%s" % (n['id'], n['score']))

        if n['score'] >= 0.66:
                print("id=>%s \tsentiment=Green \tscore=>%s" % (n['id'], n['score']))
```
